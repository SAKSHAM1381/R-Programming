# Strategy of 5 days rebalancing of buying low

```r
Total = 5000000
options(digits=9)
Remaining = Total
split = Total/5
MTM = 0
start_date <- as.Date('2018-01-02')
begin_date <- start_date
end_date <- as.Date('2018-12-31')
Full_Stock = c('IBM_AdjClose','MSFT_AdjClose','Google_AdjClose','Apple_AdjClose','Amazon_AdjClose','FB_AdjClose','NFLX_AdjClose','TSLA_AdjClose','ORCL_AdjClose','SAP_AdjClose')
Intial_Stock = c('IBM_Close','MSFT_Close','Google_Close','Apple_Close','Amazon_Close')
SharesCount = list()


while(start_date <= end_date){
    # executes the first day
    if(start_date == '2018-01-02'){
          for(i in Intial_Stock) {
        a = Stocks[Stocks$Date== start_date, Intial_Stock]
        b= floor(split/a[i])
        SharesCount[i] =  b[[i]]
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        colnames(Remaining) = 'Remaining'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        colnames(MTM) = 'MTM'
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM

    }



    else if(start_date %in% Stock_divide$Date) {

        MTM = 0
        for(i in Intial_Stock) {
        a = Stocks[Stocks$Date== start_date, Intial_Stock]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        Total = MTM + Remaining
        Remaining = Total
        split = Total / 5
        colnames(Total) = 'Total'
        colnames(MTM) = 'MTM'
        calculated_stocks = list()
        for(i in Full_Stock){

            x = Stocks[Stocks$Date== start_date, Full_Stock]
            y = Stocks[Stocks$Date== begin_date, Full_Stock]
            percent_change = 100 * ((x[i] - y[i])/ y[i])
            col = paste0(i, "_PC")
            calculated_stocks[col] = percent_change

        }
        calculated_stocks = calculated_stocks[order(unlist(calculated_stocks))]
        calculated_stocks = head(calculated_stocks,5)


        initial_col_list = names(calculated_stocks)
        print(initial_col_list)
        initial_col_list = sub('_.*','',initial_col_list)
        initial_col_list = paste0(initial_col_list,'_Close')
        IntialStocks = initial_col_list
        SharesCount = list()
        b = list()
        MTM = 0
        for(i in Intial_Stock) {
        a = Stocks[Stocks$Date== start_date, Intial_Stock]
        b= floor(split[,1]/a[i])
        SharesCount[i] =  b[[i]]
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining


    }


    else {
        MTM = 0
        for(i in Intial_Stock) {
        a = Stocks[Stocks$Date== start_date, Intial_Stock]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        colnames(MTM) = 'MTM'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM

    }
start_date = start_date+1

}
sr1_5 = c('StockSplit_5',MTM = tail(Stocks$MTM,1))
output_1 = c('Profit:', tail(Stocks$MTM,1)-5000000)
sr1_5
output_1
tail(Stocks)
```

# Starting with Time Interval of 10 Days

```r
StockDivide_10 = Stocks[seq(1,nrow(Stocks),10),]
StockDivide_10$Date = as.Date(StockDivide_10$Date)
head(StockDivide_10)
```

```r
perform_investment_strategy <- function(Total, Stocks, start_date, end_date, step = 5, StockSplit) {
  options(digits = 9)

  Remaining <- Total
  split <- Total / 5
  MTM <- 0
  InitialTotal <- Total  # Store the initial total investment

  Full_Stock <- c('IBM_AdjClose', 'MSFT_AdjClose', 'Google_AdjClose', 'Apple_AdjClose', 'Amazon_AdjClose', 'FB_AdjClose', 'NFLX_AdjClose', 'TSLA_AdjClose', 'ORCL_AdjClose', 'SAP_AdjClose')
  Intial_Stock <- c('IBM_Close', 'MSFT_Close', 'Google_Close', 'Apple_Close', 'Amazon_Close')

  SharesCount <- list()

  while (start_date <= end_date) {
    if (start_date == mdy('01/02/2018')) {
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        b <- floor(split / a[i])
        SharesCount[[i]] <- b[[i]]
        Remaining <- Remaining - b * a[i]
        MTM <- MTM + b * a[i]
      }

      colnames(Remaining) <- 'Remaining'
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
      colnames(MTM) <- 'MTM'
      MTM <- MTM + Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
    } else if (start_date %in% StockSplit$Date) {
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        MTM <- MTM + SharesCount[[i]] * a[i]
      }

      Total <- MTM + Remaining
      Remaining <- Total
      split <- Total / 5

      colnames(Total) <- 'Total'
      colnames(MTM) <- 'MTM'

      calculated_stocks <- list()
      for (i in Full_Stock) {
        x <- Stocks[Stocks$Date == start_date, Full_Stock]
        y <- Stocks[Stocks$Date == mdy('01/02/2018'), Full_Stock]
        percent_change <- 100 * ((x[i] - y[i]) / y[i])
        col <- paste0(i, "_PC")
        calculated_stocks[col] <- percent_change
      }

      calculated_stocks <- calculated_stocks[order(unlist(calculated_stocks))]
      calculated_stocks <- head(calculated_stocks, 5)

      initial_col_list <- names(calculated_stocks)
      initial_col_list <- sub('_.*', '', initial_col_list)
      initial_col_list <- paste0(initial_col_list, '_Close')
      IntialStocks <- initial_col_list

      SharesCount <- list()
      b <- list()
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        b <- floor(split[, 1] / a[i])
        SharesCount[[i]] <- b[[i]]
        Remaining <- Remaining - b * a[i]
        MTM <- MTM + b * a[i]
      }

      MTM <- MTM + Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
    } else {
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        MTM <- MTM + SharesCount[[i]] * a[i]
      }

      colnames(MTM) <- 'MTM'
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
    }

    start_date <- start_date + 1
  }
  MTM = tail(Stocks$MTM, 1)
  sr1_10 <- c('StockSplit_10', MTM )
  profit <- MTM - InitialTotal  # Calculate profit
  output_5 <- c('Profit:', profit)

  return(list(sr1_10, output_5, Stocks))
}
result <- perform_investment_strategy(5000000, Stocks, mdy('01/02/2018'), mdy('12/31/2018'), step = 10, StockDivide_10)
result
```

```r
perform_investment_strategy2 <- function(Total, Stocks, start_date, end_date, step = 5, StockSplit) {
  options(digits = 9)
  Remaining <- Total
  split <- Total / 5
  MTM <- 0
  begin_date <- mdy('01/02/2018')
  Full_Stock <- c('IBM_AdjClose','MSFT_AdjClose','Google_AdjClose','Apple_AdjClose','Amazon_AdjClose','FB_AdjClose','NFLX_AdjClose','TSLA_AdjClose','ORCL_AdjClose','SAP_AdjClose')
  Intial_Stock <- c('IBM_Close','MSFT_Close','Google_Close','Apple_Close','Amazon_Close')
  SharesCount <- list()

  while (start_date <= end_date) {
    if (start_date == begin_date) {
      for (i in IntialStocks) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        b <- floor(split / a[i])
        SharesCount[[i]] <- b[[i]]
        Remaining <- Remaining - b * a[i]
        MTM <- MTM + b * a[i]
      }
      colnames(Remaining) <- 'Remaining'
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
      colnames(MTM) <- 'MTM'
      MTM <- MTM + Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
    }
    else if (start_date %in% StockSplit$Date) {
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        MTM <- MTM + SharesCount[[i]] * a[i]
      }
      Total <- MTM + Remaining
      Remaining <- Total
      split <- Total / 5
      colnames(Total) <- 'Total'
      colnames(MTM) <- 'MTM'
      calculated_stocks <- list()
      for (i in Full_Stock) {
        x <- Stocks[Stocks$Date == start_date, Full_Stock]
        y <- Stocks[Stocks$Date == begin_date, Full_Stock]
        percent_change <- 100 * ((x[i] - y[i]) / y[i])
        col <- paste0(i, "_PC")
        calculated_stocks[col] <- percent_change
      }
      calculated_stocks <- calculated_stocks[order(unlist(calculated_stocks))]
      calculated_stocks <- tail(calculated_stocks, 5)
      initial_col_list <- names(calculated_stocks)
      initial_col_list <- sub('_.*', '', initial_col_list)
      initial_col_list <- paste0(initial_col_list, '_Close')
      IntialStocks <- initial_col_list
      SharesCount <- list()
      b <- list()
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        b <- floor(split[, 1] / a[i])
        SharesCount[[i]] <- b[[i]]
        Remaining <- Remaining - b * a[i]
        MTM <- MTM + b * a[i]
      }
      MTM <- MTM + Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
    }
    else {
      MTM <- 0
      for (i in Intial_Stock) {
        a <- Stocks[Stocks$Date == start_date, Intial_Stock]
        MTM <- MTM + SharesCount[[i]] * a[i]
      }
      colnames(MTM) <- 'MTM'
      Stocks[Stocks$Date == start_date, 'Remaining'] <- Remaining
      Stocks[Stocks$Date == start_date, 'MTM'] <- MTM
    }
    start_date <- start_date + 1
  }
    MTM = tail(Stocks$MTM, 1)
  sr2_7 <- c('MTM', MTM )
  output_7 <- c('Loss:', 5000000 - MTM)
  return(list(sr2_7, output_7, tail(Stocks)))
}

# Example usage
result23 <- perform_investment_strategy2(5000000, Stocks, mdy('01/02/2018'), mdy('12/31/2018'), 7, StockDivide_7)
result23
```
